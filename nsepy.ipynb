```python
# we will get warning if we remove loc and row part directing asking columns
dfreg = sbin.loc[:,['Close','Volume']]

dfreg['HL_PCT'] = (sbin['High'] - sbin['Low']) / sbin['Close'] * 100.0
dfreg['PCT_change'] = (sbin['Close'] - sbin['Open']) / sbin['Open'] * 100
dfreg
```

```python
# Drop missing value
dfreg.fillna(value=-99999, inplace=True)
```

```python
import math
# We want to separate 1 percent of the data to forecast
forecast_out = int(math.ceil(0.01 * len(dfreg)))
```

```python
# Separating the label here, we want to predict the Close
forecast_col = 'Close'
dfreg['label'] = dfreg[forecast_col].shift(-forecast_out)
```

```python
# from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn import preprocessing
```
